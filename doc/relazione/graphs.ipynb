```python
import csv
from os import listdir
from os.path import isfile, join
import matplotlib.pyplot as plt
import matplotlib

mypath = '../results/'
```

```python
def all_graphs():
    files = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    for file_name in files:
        sizes = []
        times = []
        gflops = []
        with open(join(mypath, file_name), 'rb') as csvfile:
            reader = csv.reader(csvfile, delimiter='|')
            for row in reader:
                size = int(row[0])
                sizes.append(size)
                time = float(row[1])
                times.append(time)
                gflop = float(row[2])
                gflops.append(gflop)

            plt.figure("{}_1".format(file_name))
            plt.title(file_name)
            plt.xlabel('size')
            plt.ylabel('seconds')
            plt.grid(True)
            plt.legend()
            plt.plot(sizes, times, lw=2, marker='o', label=file_name)

            plt.figure("{}_2".format(file_name))
            plt.title(file_name)
            plt.xlabel('size')
            plt.ylabel('gflop/seconds')
            plt.grid(True)
            plt.legend()
            plt.plot(sizes, gflops, lw=2, color='r', marker='o', label=file_name)
    plt.show()
```

```python
def blocking_graphs_times():
    files = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    plt.title("blocking experiments")
    plt.xlabel('size')
    plt.ylabel('seconds')
    plt.grid(True)
    
    legend_items = []
    for file_name in files:
        if 'nonblock' not in file_name:
            sizes = []
            times = []
            gflops = []
            with open(join(mypath, file_name), 'rb') as csvfile:
                reader = csv.reader(csvfile, delimiter='|')
                for row in reader:
                    size = int(row[0])
                    sizes.append(size)
                    time = float(row[1])
                    times.append(time)
                    gflop = float(row[2])
                    gflops.append(gflop)
            plt.plot(sizes, times, lw=2, marker='o', label=file_name)
    
    plt.legend(loc=2)
    plt.show()

def blocking_graphs_gflops():
    files = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    plt.title("blocking experiments")
    plt.xlabel('size')
    plt.ylabel('gflop/s')
    plt.grid(True)
    
    legend_items = []
    for file_name in files:
        if 'nonblock' not in file_name:
            sizes = []
            times = []
            gflops = []
            with open(join(mypath, file_name), 'rb') as csvfile:
                reader = csv.reader(csvfile, delimiter='|')
                for row in reader:
                    size = int(row[0])
                    sizes.append(size)
                    time = float(row[1])
                    times.append(time)
                    gflop = float(row[2])
                    gflops.append(gflop)
            plt.plot(sizes, gflops, lw=2, marker='o', label=file_name)
    
    plt.legend(loc=2)
    plt.show()
            
blocking_graphs_times()
blocking_graphs_gflops()
```
